```python
from bisect import bisect_left
def binSearch(a, x):
    i= bisect_left(a, x)
    if i != len(a) and a[i] == x:
        return i
    else:
        return -1
```

```python
def searchFromIn(sublist,orderedList):
    def idxInOrderedListOfItemInSublist(sublistIdx):
        element=sublist[sublistIdx]
        return binSearch(orderedList, element)
    return idxInOrderedListOfItemInSublist
```

```python
def valueInList(searchList):
    def returnValue(idx):
        return searchList[idx]
    return returnValue
```

```python
import math
def generateArrangementNumber(sublist, totalList):
    totalList=sorted(totalList)
    arrNumber=0
    #Sum(i=0:V-1|N!/(N-i)!+(V-i)!(ji-sum(k=0:i|-11{xk<xji})))
    N=len(totalList)
    V=len(sublist)
    fact=math.factorial
    j=searchFromIn(sublist,totalList)
    x=valueInList(totalList)
    arrNumber=sum([fact(N)/fact(N-i)+fact(V-i)*(j(i)-sum([1 for k in range(i) if x(k)<x(j(i))])) for i in range(V)])
    return arrNumber
```

```python
import math
def generatePermutationNumber(totalList):
    arrNumber=0
    #Sum(i=0:V-1|i!(sum(k=0:i|-1{xk<xi})))
    N=len(totalList)
    fact=math.factorial
    x=valueInList(totalList)
    
    arrNumber=sum([fact(N-1-i)*sum(1 for k in range(i+1,N) if x(i)>x(k)) for i in range(0,N-1)])
    return arrNumber
```

```python
for i in range(len(big)+1):
    for perm in permutations(big,i):
        print(perm)
        print(generatePermutationNumber(perm))
```

```python
def nbCombi(toFind,among):
    fact=math.factorial
    return fact(among)/(fact(among-toFind)*fact(toFind))
```
